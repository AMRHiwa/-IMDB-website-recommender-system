## مدل مبتنی بر محتوا

برای اینکار مقادیر موجود در ستون genres را به لیستی از ژانرها تقسیم کنیم تا برای استفاده در آینده ساده شود.میتوانیم با اعمال تابع split با مقدار ' | ' در ستون  صحیح به دست آورد.

```python
df_movie['genres'] = df_movie.genres.str.split('|')
df_movie.head()
```

از آنجایی که حفظ ژانرها در قالب لیست برای تکنیک سیستم توصیه مبتنی بر محتوا بهینه نیست، از تکنیک One Hot Encoding برای تبدیل لیست ژانرها به بردار استفاده می کنیم که در آن هر ستون با یک مقدار ممکن از ویژگی مطابقت دارد. این رمزگذاری برای تغذیه داده های طبقه بندی مورد نیاز است. در این مورد، ما هر ژانر مختلف را در ستون‌هایی ذخیره می‌کنیم که دارای 1 یا 0 هستند. بیایید این چارچوب داده را در متغیر دیگری نیز ذخیره کنیم زیرا ژانرها برای اولین سیستم توصیه ما مهم نخواهند بود.

برای این کار یک دیتا فریم جدید کپی از دیتا فریم قبلی ایجاد میکنیم و در ژانر های متفاوت را در داخل آن ذخیره میکنیم. 

با استفاده از یک حلقه ، سعی میکنیم در ازای حرکت در هر ردیف یک اندیس ، برای هر یک از ژانر های موجود در لیست در ستون genres یک ستون با نام آن ژانر ایجاد کرده و مقدار 1 را به آن نسبت میدهد. اینکار را برای تمام اندیس ها یا به عبارتی برای تمام سطرها انجام میدهد تا اینکه تمام ژانر ها دارای ستون شوند. مرحله بعد جایگذاری مقادیر تعریف نشده با مقدار 0 است چراکه ستون های که بعدا اضافه میشوند ابتدا در سطر های آغازین نبودند و مقدار not availble رو میگیرند.

```python
moviesWithGenres_df = df_movie.copy()

for index, row in df_movie.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = 1

moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df.head()
```

حال کار ما با دیتاست فیلم ها پایان یافت ، باید به سراغ مرحله بعدی یعنی دیتاست raiting بریم

```python
df_raiting.head()
```

در دیتاست raiting نیاز چندانی به ستون timestamp نیست پس اون رو از دیتاست حذف میکنیم چراکه هدف ما پیدا کردن نظر کاربران به فیلم هایی دیدن است 

```python
df_raiting = df_raiting.drop('timestamp', axis= 1)
df_raiting.head()
```

یک کاربر برای خودمون در نظر میگیریم که فیلم های زیر را دیده و امتیاز های زیر رو به اون ها داده حال این داده ها را در دیتافریم ذخیره میکنیم 

```python
userInput = [
            {'title':'Heat', 'rating':4},
            {'title':'Toy Story', 'rating':4},
            {'title':'Mortal Kombat', 'rating':3},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5},
            {'title':'Bad Boys', 'rating': 5},
            {'title':'Batman Forever', 'rating':3.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

**افزودن movieId به کاربر ورودی**
- با ایجاد یک دیتافریم ورودی برای یک کاربر(مثلا)حال نوبت آن فرا میرسد که شناسه فیلم ورودی را از دیتا فریم فیلم استخراج کرده و به آن اضافه کنیم.

- ما می توانیم با فیلتر کردن ردیف هایی که حاوی عنوان فیلم ورودی هستند و سپس ادغام این زیرمجموعه با دیتافریم ورودی به این هدف دست یابیم. همچنین برای صرفه جویی در فضای حافظه، ستون های غیر ضروری را برای ورودی رها می کنیم.

#### نکته
-اگر فیلمی که در بالا اضافه کردیم اینجا نیست، ممکن است در نسخه اصلی نباشد یا ممکن است املای آن متفاوت باشد، لطفاً حروف بزرگ را باید بررسی کنیم.

1. ابتدا فیلم ها را بر اساس نام آنها در جدول df_movie پیدا کرده. 
2. سپس آن را ادغام میکنیم تا بتوانیم شناسه فیلم را دریافت کنیم.
3. رهاسازی اطلاعاتی که از آنها در دیتافریم ورودی استفاده نمیکنیم ( ستون های year و genres )
4. نمایش دیتا فریم ادغامی ایجاد شده 

```python
# قدم اول
inputId = df_movie[df_movie['title'].isin(inputMovies['title'].tolist())]
# قدم دوم
inputMovies = pd.merge(inputId, inputMovies)
# قدم سوم
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)

inputMovies
```

* حال در این مرحله میخواهیم ژانر فیلم هایی را که ایدی آنها در دیتافریم movieWithGenres_df موجود است را به دیتافریم خود اضافه کنیم.
* برای اینکار ما یک کپی از دیتا فریم movieWithGenres_df ایجاد میکنیم اما تنها با مقداری که در آیدی آنها در دیتافریم inputMovies موجود بوده است. 

```python
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

حال پس از به دست آوردن اطلاعات مورد نیاز درون یک دیتا فریم نیاز است که داده ها اضافی حذف شوند:
- پس از دریافت داده ها مورد نیاز جدول را برای اندیس ها ری ست میکنیم تا از اول به همه یک اندیس دهد.
- ما قصد داریم که به ژانر ها بر اساس امتیاز کاربر یکسری امتیاز بدهیم پس دیگر نیازی به ستون های دیگر چون عنوان و ژانر و ... نداریم 

```python
# گام اول
userMovies = userMovies.reset_index(drop=True)
# گام دوم
userGenreTable = userMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
userGenreTable
```

***شروع یادگیری:***

برای انجام این کار، ما قصد داریم هر ژانر را به وزن تبدیل کنیم. ما می توانیم این کار را با استفاده از بررسی های ورودی و ضرب آنها در جدول ژانر ورودی و سپس جمع بندی جدول به ستون به دست آمده انجام دهیم. این عملیات در واقع  یک ماتریس و حاصل ضرب داخلی یک بردار است، بنابراین ما می توانیم به سادگی با فراخوانی تابع dot برای ضرب داخلی ماتریس ها در ماژول pandas این کار را انجام دهیم.

```python
inputMovies['rating']
```

برای اینکار کافی است که مقدار بردار rating را در هر سطر ماتریس inputMovies ضرب شده تا ضریب هر کدام بدست بیاید و به طور ستونی حاصل جمع هر ستون رو محاسبه میکنیم

```python
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
userProfile
```

حال ضریب هر کدام از ویژگی های محبوب کاربر را داریم و با این کار میتوانیم عناوینی را به کار توصیه کنیم که دارای ضریب بیشتری هستند. حال برای اینکار شروع به استخراج جدول genres از دیتافریم اصلی میکنیم و داده های زايد را از جدول حذف میکنیم:

```python
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])

genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
genreTable.head()
```

```python
genreTable.shape
```

با در دست داشتن مشخصات ورودی و فهرست کامل فیلم‌ها و ژانرهای آن‌ها، میانگین وزنی هر فیلم را بر اساس نمایه ورودی دریافت می‌کنیم و بیست فیلم برتر را پیشنهاد می‌کنیم که بیشترین رضایت را دارند.

```python
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

بعد از به دست آوردن میانگین وزنی ما باید این داده ها را به صورت نزولی از بالا به پایین مرتب کنیم و ارزش هرکدام از فیلم ها رو برای خودمون نمایش میدیم مثلا 15 فیلم اول بنا به ارزش دهی 

```python
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
recommendationTable_df.head(15)
```

بعد از این محاسبه و به دست اوردن ارزش هریک از داده های جدول ما میتوانیم یک جدول برای پیشنهادات خودمون ایجاد کنیم که برای اینکار کافی است MovieId فیلم ها رو در دیتا فریم اصلی پیدا کنیم و در جدول پیشنهادات اضافه کنیم.

```python
df_movie.loc[df_movie['movieId'].isin(recommendationTable_df.head(15).keys())]
```

```python
userInput = [
            {'title':'Heat', 'rating':4},
            {'title':'Toy Story', 'rating':4},
            {'title':'Mortal Kombat', 'rating':3},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5},
            {'title':'Bad Boys', 'rating': 5},
            {'title':'Batman Forever', 'rating':3.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

فیلتر کردن کاربرانی که فیلم هایی را که ورودی تماشا کرده اند و ذخیره آن

Groupby چندین فریم داده فرعی ایجاد می کند که همه آنها دارای یک مقدار در ستون مشخص شده به عنوان پارامتر هستند

به عنوان مثال ما 6 نفر اول این لیست را مشاهده میکنیم.

```python
userSubsetGroup = sorted(userSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
userSubsetGroup[0:6]
```

حال نوبت بررسی شباهت کاربران با کاربر اصلی است.

در این مرحله میخوایم تعدادی از کاربران را با کاربر مشخص شده خود مقایسه کنیم و یکی را پیدا کنیم که بیشتر شباهت را به آن داشته باشد.
از طریق ضریب همبستگی پیرسون میتوانیم بفهمیم که هر کاربر چقدر به کاربر اصلی شباهت دارد. که در زیر برای شما دلیل استفاده از آن را بیان میکنم

دلیل استفاده از همبستگی پیرسون این است که :
همبستگی پیرسون نسبت به مقیاس بندی ثابت است، یعنی ضرب همه عناصر در یک ثابت غیر صفر یا اضافه کردن هر ثابتی به همه عناصر. به عنوان مثال، اگر دو بردار X و Y دارید، پس پیرسون (X, Y) == pearson (X, 2 * Y + 3). این یک ویژگی بسیار مهم در سیستم‌های توصیه است، زیرا برای مثال دو کاربر ممکن است دو سری از آیتم‌ها را کاملاً متفاوت از نظر نرخ‌های مطلق رتبه‌بندی کنند، اما کاربران مشابه (یعنی با ایده‌های مشابه) با نرخ‌های مشابه در مقیاس‌های مختلف باشند.

![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/bd1ccc2979b0fd1c1aec96e386f686ae874f9ec0 "Pearson Correlation")


مقادیر داده شده توسط فرمول از r = -1 تا r = 1 متفاوت است، که در آن 1 یک همبستگی مستقیم بین دو موجودیت را تشکیل می دهد (به معنای همبستگی مثبت کامل است) و -1 یک همبستگی منفی کامل را تشکیل می دهد.در این جا 1 به این معنی است که این دو کاربر دارای سلیقه مشابه هستند در حالی که -1 به معنای مخالف است.زیر مجموعه ای از کاربران را برای تکرار انتخاب می کنیم چرا که نمیخواهیم زمان بیشتری رو صرف محاسبه انجام بدیم

```python
userSubsetGroup = userSubsetGroup[0:100]
```

همبستگی پیرسون را بین کاربر ورودی و گروه زیرمجموعه محاسبه کرده در یک dictionary ذخیره می‌کنیم که key آن UserID و Value آن ضریب است.

```python
pearsonCorrelationDict = {}
for name, group in userSubsetGroup:
    group = group.sort_values(by='movieId')
    inputMovies = inputMovies.sort_values(by='movieId')
    nRatings = len(group)
    temp_df = inputMovies[inputMovies['movieId'].isin(group['movieId'].tolist())]
    tempRatingList = temp_df['rating'].tolist()
    tempGroupList = group['rating'].tolist()
    Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/np.sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```

```python
for i in pearsonCorrelationDict.items():
    print(i)
```

```python
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
pearsonDF.columns = ['similarityIndex']
pearsonDF['userId'] = pearsonDF.index
pearsonDF.index = range(len(pearsonDF))
pearsonDF.head(15)
```

حالا 100 کاربر برتر را که بیشترین شباهت را به ورودی دارند، پیدا میکنیم.

```python
topUsers=pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:100]
topUsers.head(15)
```
